## Summary in single column

```python
def data_details(data):
    print("-----Shape-----")
    print(data.shape)
    
    print("-----info-----")
    print(data.info())
    
    print("----description-----")
    print(data.describe())
    
    print("-----duplicate-----")
    print(data.duplicated())
    
    print("-----missing values-----")
    print(data.isnull().sum())
```

```python
data_details(df)
```

## CUT & QCUT

```python
df.head()
```

```python
df["Age_Class"]=pd.qcut(df["Age"],q=4)
```

```python
df[["Age","Age_Class"]]
```

```python
df["Education"].value_counts()
```

```python
df["Education_Classes"]=pd.qcut(df["Education"],q=2,labels=["poor","good"])
```

```python
df[["Education","Education_Classes"]]
```

```python
df['DistanceFromHome'].value_counts()
```

```python
df['commute']=pd.qcut(df['DistanceFromHome'],q=3, labels = ["near", "far", "most far"])
```

```python
df[['DistanceFromHome','commute']].value_counts()
```

```python
df['cut_age']=pd.cut(df['Age'],bins = [18,30,40,50,60], labels = ["18-29","30-39","40-49","50-59"])
```

```python
df[['Age','cut_age']]
```
